```python
# Correct import
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain_ollama.llms import OllamaLLM
```

```python
template = """ context: {input_text}

summarize the above text in 50 words """
```

```python
template
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
prompt
```

```python
model = OllamaLLM(model = "llama3:latest")
```

```python
chain = prompt | model
```

```python
input = """LangChain is an open-source orchestration framework designed to simplify the creation of applications using large language models (LLMs). It provides a standardized interface for interacting with different models, abstracting away the complexities of API management and allowing developers to seamlessly swap providers like OpenAI, Anthropic, or Google. The framework is available as libraries in both Python and JavaScript.
The core principle of LangChain is "chaining" together different components to create cohesive, multi-step workflows. This modular approach enables complex language-based AI systems to be built from reusable parts.
Key Concepts & Components
Models: LangChain provides a standard interface (APIs) for connecting to various public and proprietary LLMs, enabling developers to interact with models using simple, consistent code.
Prompt Templates: These are pre-built, reusable structures used to format prompts for AI models consistently and precisely, facilitating effective prompt engineering and few-shot learning.
Chains: Chains are sequences of actions that combine LLMs with other components, executing a series of functions to achieve a specific goal. The simplest form is an LLMChain, which links a model and a prompt template.
Agents: Agents are dynamic components that use an LLM as a "reasoning engine" to decide the best sequence of actions (using tools) in response to a user query. This gives the AI the ability to make decisions and perform complex tasks autonomously.
Tools: These are functions that agents can call to interact with external systems, such as search engines (e.g., Google Search), databases (e.g., SQL or vector stores), or other APIs (e.g., Wolfram Alpha, weather services).
Retrieval: This module is essential for Retrieval-Augmented Generation (RAG) systems, allowing LLMs to access new, domain-specific data without retraining, reducing model hallucinations and improving accuracy. Components include:
Document Loaders: For importing data from various sources (PDFs, websites, databases).
Text Splitters: To break large documents into smaller, semantically meaningful chunks for efficient processing.
Embeddings: To convert text chunks into numerical vector representations.
Vector Stores: Specialized databases (e.g., Chroma, Milvus) for storing and performing similarity searches on these vectors.
Retrievers: To fetch the most relevant data chunks based on a user's query."""
```

```python
input_user = {"input_text": input}
```

```python
chain.invoke(input_user)
```
